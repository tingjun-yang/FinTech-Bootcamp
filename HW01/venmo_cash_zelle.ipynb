```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(1,2,figsize=(12, 4),dpi=200)

venmo_revenue = np.array([160, 200, 300, 450,850, 935])
venmo_revenue = venmo_revenue*1e-3
venmo_revenue_year = np.array([2017, 2018, 2019, 2020, 2021, 2022])

cash_revenue = np.array([0.4, 1.3, 5.9, 12.3])
cash_revenue_year = np.array([2018, 2019, 2020, 2021])

axs[0].scatter(venmo_revenue_year, venmo_revenue, s=20, c='b', marker="s", label='Venmo')
axs[0].scatter(cash_revenue_year, cash_revenue, s=20, c='r', marker="o", label='Cash App')
axs[0].set_title("Annual Revenue")
axs[0].set_xlabel("Year")
axs[0].set_ylabel("Annual Revenue ($bn)")
axs[0].legend(loc='upper left')


venmo_payment_volume = [20, 34, 62, 101, 159, 230]
cash_payment_volume = [33, 50, 65, 85, 112, 167]
zelle_payment_volume = [66, 75, 119, 187, 307, 490]
payment_volume_year = range(2016, 2022)

axs[1].scatter(payment_volume_year, venmo_payment_volume, s=20, c='b', marker="s", label='Venmo')
axs[1].scatter(payment_volume_year, cash_payment_volume, s=20, c='r', marker="o", label='Cash App')
axs[1].scatter(payment_volume_year, zelle_payment_volume, s=20, c='g', marker="^", label='Zelle')
axs[1].set_title("Annual Payment Volume")
axs[1].set_xlabel("Year")
axs[1].set_ylabel("Annual Payment Volume ($bn)")
axs[1].legend(loc='upper left')

#plt.show()
plt.savefig("venmo.png")
```
